# Remove Punctuation

```python
import string
```

```python
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text = text.replace(char, "")
    return text
```

```python
df['title'] = df['title'].apply(remove_punc)
```

```python
df['overview'] = df['overview'].apply(remove_punc)
```

```python
df['genre_names'] = df['genre_names'].apply(remove_punc)
```

```python
df['genre_names'] = df['genre_names'].apply(lambda genres: ''.join(genres))
```

```python
df
```

# Remove Stop Words

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    new_txt = []
    for word in text.split():
        if word in stopwords.words('english'):
            new_txt.append('')
        else:
            new_txt.append(word)
    x = new_txt[:]
    new_txt.clear()
    return " ".join(x)
```

```python
df['title'] = df['title'].apply(remove_stopwords)
```

```python
df['overview'] = df['overview'].apply(remove_stopwords)
```

```python
df['genre_names'] = df['genre_names'].apply(remove_stopwords)
```

```python
df
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem_word(text):
    return " ".join([ps.stem(word) for word in text])
```

```python
df['title'] = df['title'].apply(stem_word)
```

```python
df['overview'] = df['overview'].apply(stem_word)
```

```python
df['genre_names'] = df['genre_names'].apply(stem_word)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
exclude = string.punctuation
```

```python
def lemmatize_overview(text):
    return ' '.join([wordnet_lemmatizer.lemmatize(word, pos='v') for word in text.split()])
```

```python
df['overview'] = df['overview'].apply(lemmatize_overview)
```
