```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
```

```python
df = pd.read_csv('update_movies.csv')
```

```python
df
```

```python
def remove_stopwords(text):
    new_txt = []
    for word in text.split():
        if word in stopwords.words('english'):
            new_txt.append('')
        else:
            new_txt.append(word)
    x = new_txt[:]
    new_txt.clear()
    return " ".join(x)
```

```python
df['title'] = df['title'].apply(remove_stopwords)
```

```python
df['overview'] = df['overview'].apply(remove_stopwords)
```

```python
df['genre_names'] = df['genre_names'].apply(remove_stopwords)
```
